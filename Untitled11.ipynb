```python

num = int(input("Enter a number: "))

if num <= 1:
    print("The number is not a prime number",num)
else:
    for i in range(2, num):
        if num % i == 0:
            print("The number is not a prime number",num)
            break
    else:
        print("The number is a prime number",num)
```
